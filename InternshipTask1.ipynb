```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset from seaborn
tips = sns.load_dataset('tips')

# Information about the dataset
print(tips.info())

# Set up the figure and axes
fig, axes = plt.subplots(3, 2, figsize=(16, 18))

# 1. Count Plot for Gender Distribution
sns.countplot(x='sex', data=tips, palette='Set2', ax=axes[0, 0])
axes[0, 0].set_title('Distribution of Gender')
axes[0, 0].set_xlabel('Gender')
axes[0, 0].set_ylabel('Count')

# 2. Pie Chart for Smoker Status
smoker_counts = tips['smoker'].value_counts()
axes[0, 1].pie(smoker_counts, labels=smoker_counts.index, autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#99ff99'])
axes[0, 1].set_title('Distribution of Smoker Status')

# 3. Bar Plot for Day of the Week
sns.barplot(x='day', y='total_bill', data=tips, ci=None, palette='Set3', ax=axes[1, 0])
axes[1, 0].set_title('Average Total Bill by Day')
axes[1, 0].set_xlabel('Day')
axes[1, 0].set_ylabel('Average Total Bill')

# 4. Box Plot for Time of Day vs. Total Bill
sns.boxplot(x='time', y='total_bill', data=tips, palette='Set1', ax=axes[1, 1])
axes[1, 1].set_title('Total Bill Distribution by Time of Day')
axes[1, 1].set_xlabel('Time of Day')
axes[1, 1].set_ylabel('Total Bill')

# 5. Histogram for Total Bill Distribution
sns.histplot(tips['total_bill'], bins=20, color='skyblue', edgecolor='black', ax=axes[2, 0])
axes[2, 0].set_title('Histogram of Total Bill Distribution')
axes[2, 0].set_xlabel('Total Bill')
axes[2, 0].set_ylabel('Frequency')

# Remove the unused subplot
fig.delaxes(axes[2, 1])

# Adjust layout
plt.tight_layout()

# Save the figure
plt.savefig('variety_of_categorical_charts.png')

# Show the plots
plt.show()
```
